```python
def process_url(url):
    
    p = []
    max_page = 1
    
    while True:
        try:
            r = requests.get(url, headers=headers)
            break
        except Exception as e:
            print(e)
            time.sleep(5.0)

    
    try: 
        soup = BeautifulSoup(r.text, "lxml")
        a = soup.find('div',attrs = {"class":"pagination"})
        a = [a_.text for a_ in a.find_all('a',attrs = {"class":"button"})]
        pgno = [int(x) for x in a if len(x) == 1]

        if pgno:
            max_page = max(pgno)
        topic_title = soup.find('h2', class_="topic-title").text
        
    except Exception as e:
        print('error in base link:', url)
        print(e)
        return None
    
    for x in range(0,max_page*10,10):

        while True:
            try:
                url_ = url.replace(".html", "-s" + str(x) + ".html")
                r = requests.get(url_, headers=headers)

                break
            except Exception as e:
                print(e)
                time.sleep(5.0)

        soup = BeautifulSoup(r.text, "lxml")

        try:
            h = soup.find_all('div', attrs = {"class":"postbody"})
            for x in h:
                post = x.find('div', attrs={"class": "pb2 postcontentheight postcontentheight1"})
                if post:
                    content = [p.text for p in post.find_all() if len(p.text.split()) > 5]
                    p.extend(content)
        except Exception as e:
            print('error in content post:', url_)
            print(e)
            return
    
    time.sleep(8)
    if p:
        data = {'url': url, 'topic_title': topic_title, 'content': p}
    else:
        return
    
    return data
```
