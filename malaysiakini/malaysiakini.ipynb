```python
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
import pickle
import json
import requests
from concurrent.futures import ThreadPoolExecutor
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.options import Options
```

```python
links = [
"https://www.malaysiakini.com/my/latest/columns",
"https://www.malaysiakini.com/my/latest/letters",
"https://www.malaysiakini.com/my/tag/ulasan",
"https://www.malaysiakini.com/my/announcement",
"https://www.malaysiakini.com/my/tag/global",
"https://www.malaysiakini.com/my/latest/sukan",
"https://www.malaysiakini.com/my/latest/news",
"https://www.malaysiakini.com/my/latest/hiburan"
]

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} 

options = Options()
options.add_argument("start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
```

```python
def save_cookie(driver, path):
    with open(path, 'wb') as filehandler:
        pickle.dump(driver.get_cookies(), filehandler)
```

```python
driver = webdriver.Edge( options=options)

# We open the website before loading the cookies into the driver
# to avoid the InvalidCookieDomainException
driver.get('https://www.malaysiakini.com/my/latest/hiburan?page=2')

save_cookie(driver,'cookies.pkl')

cookies = pickle.load(open("cookies.pkl", "rb"))
```

```python
hrefs = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} 
driver = webdriver.Edge(options=options)
for i in tqdm(links):

    x = 1

    while True:

        while True:
            try:
                driver.get(f'{i}?page={x}')
                for cookie in cookies:
                    driver.add_cookie(cookie)
                # Wait for the page to load completely
                print(f'{i}?page={x}')
                wait = WebDriverWait(driver, 10)
                page = driver.page_source
                
                break
            except Exception as e:
                print(e)
                time.sleep(1.0)
                
        soup = BeautifulSoup(page, "html.parser")

        div = soup.find_all('div',attrs = {"class":"cursor-pointer overflow-hidden shadow-lg text-coolGray-600 group relative h-full mb-4"})

        if len(div) == 0:
            break
        elif div[0].find('a').get('href') == r'/[lang]/latest/function%20link()%20%7B%20[native%20code]%20%7D' or div[0].find('a').get('href') == r'/[lang]/tag/function%20link()%20%7B%20[native%20code]%20%7D':
            break


        for link in soup.find_all('div',attrs = {"class":"cursor-pointer overflow-hidden shadow-lg text-coolGray-600 group relative h-full mb-4"}):
            
            href = link.find('a').get('href')
            print(href)

           
            hrefs.append('https://www.malaysiakini.com'+href)
                
        
        x = x + 1
```

```python
def process_url(x):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'} # This is chrome, you can set whatever browser you like

    
    while True:
        try:
            r = requests.get(x, headers=headers)
            break
        except Exception as e:
            print(e)
            time.sleep(5.0)
    
    soup = BeautifulSoup(r.text, "lxml")

    try:
        headline = soup.find('div', attrs = {"class":"text-3xl font-semibold leading-snug text-coolGray-600 mt-2 tracking-normal print:text-lg"}).text
        h = soup.find('div', attrs = {"class":"px-4 lg:px-0"})        
        p = [p.text for p in h.find_all()]
    except Exception as e:
        print('error in link:'+ x)
        print(e)
        return None 
    
    
    data = {'url': x, 'headline': headline, 'content': p}
    return data
```

```python
max_workers = 30

with ThreadPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(process_url, x) for x in hrefs]
    
    for future in tqdm(futures, total=len(hrefs)):
        result = future.result()
        if result:
            with open('malaysiakini-bm.jsonl', 'a') as f:
                json.dump(result, f)
                f.write('\n')
```
